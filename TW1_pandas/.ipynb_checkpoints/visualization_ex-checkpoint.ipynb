### Why data visualization?
- Gain insight into an information space by mapping data onto graphical primitives

- Provide qualitative overview of large data sets

- Search for patterns, trends, structure, irregularities, relationships among data

- Help find interesting regions and suitable parameters for further quantitative analysis

- Provide a visual proof of computer representations derived


```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_all = pd.read_csv('data.csv')

df_all.head(5)
```

```python
df_all.describe()
```

```python
df = df_all[['bullseyeX', 'bullseyeY', 'circleX', 'circleY', 'starX', 'starY', 'dinoX', 'dinoY']]
df.head()
```

```python
fig, ((ax1, ax2)) = plt.subplots(1, 2, sharey = True)

ax1.boxplot([df['bullseyeX'], df['circleX'], df['starX'], df['dinoX']])
ax2.boxplot([df['bullseyeY'], df['circleY'], df['starY'], df['dinoY']])

fig.suptitle('Box plots')
ax1.set_title('X boxplots')
ax2.set_title('Y boxplots')

plt.show()
```

### Another way to make subplots

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

fig.suptitle('Box plots')
ax1.set_title('X boxplots')
ax2.set_title('Y boxplots')

ax1.boxplot([df['bullseyeX'], df['circleX'], df['starX'], df['dinoX']])
ax2.boxplot([df['bullseyeY'], df['circleY'], df['starY'], df['dinoY']])

plt.show()
```

```python
fig, ((ax1, ax2, ax3, ax4), 
      (ay1, ay2, ay3, ay4)) = plt.subplots(2, 4, figsize = [8, 6], 
                                           sharey=True, sharex=True)

ax1.hist(df['bullseyeX'])
ax2.hist(df['circleX'])
ax3.hist(df['starX'])
ax4.hist(df['dinoX'])

ay1.hist(df['bullseyeY'])
ay2.hist(df['circleY'])
ay3.hist(df['starY'])
ay4.hist(df['dinoY'])

fig.suptitle('Histotrams')

plt.show()
```

```python
fig, ((a1, a2), (a3, a4)) = plt.subplots(2, 2, figsize = [8, 8],
                                        sharex=True, sharey=True)

a1.scatter(df['bullseyeX'], df['bullseyeY'])
a2.scatter(df['circleX'], df['circleY'])
a3.scatter(df['starX'], df['starY'])
a4.scatter(df['dinoX'], df['dinoY'])

a1.set_ylabel('Y')
a3.set_xlabel('X')
a3.set_ylabel('Y')
a4.set_xlabel('X')

fig.suptitle('Scatter plots')

plt.show()
```
